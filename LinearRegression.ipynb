```python
import pandas as pd
df = pd.DataFrame({'x':[0,1,2,3,4],'y':[1,3,5,7,9]})
```

```python
df.head()
```

```python
x = df[['x']]
y = df['y']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
print(f'f(x) = {int(model.coef_[0])} * x+{int(model.intercept_)}')
```

```python
model.predict([[5]])
```
